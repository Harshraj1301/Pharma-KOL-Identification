```python
# Create a new DataFrame by repeating each row for each author and assigning the individual author to 'Single Author'
expanded_rows = []  # To collect all the expanded rows with individual authors
for _, row in Research_Journal.iterrows():
    authors = row['Authors'].split(', ')
    for author in authors:
        new_row = row.copy()
        new_row['Single Author'] = author
        expanded_rows.append(new_row)

# Convert the list of series into a DataFrame
expanded_Research_Journal = pd.DataFrame(expanded_rows)

# If needed, reset the index of the new DataFrame
expanded_Research_Journal.reset_index(drop=True, inplace=True)

# Display the first few rows to verify the output
print(expanded_Research_Journal[['PMID', 'Title', 'Single Author','First Author']].head(10))
```

```python
expanded_Research_Journal['Is Lead Author'] = (expanded_Research_Journal['Single Author'] == expanded_Research_Journal['First Author']).astype(int)
print(expanded_Research_Journal)
```

```python
# Group by 'Single Author' and aggregate data
author_level_aggregate = expanded_Research_Journal.groupby('Single Author').agg(
    number_of_publications=pd.NamedAgg(column='PMID', aggfunc='nunique'),
    number_of_citations=pd.NamedAgg(column='DOI', aggfunc='nunique'),
    number_of_times_lead=pd.NamedAgg(column='Is Lead Author', aggfunc='sum')
).reset_index()

# Display the aggregated DataFrame
print(author_level_aggregate)
```
